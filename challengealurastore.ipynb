```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un DataFrame con datos ficticios (puedes ajustar estos valores o cargar tus propios datos)
data = {
    'tienda': ['Tienda A', 'Tienda A', 'Tienda B', 'Tienda B', 'Tienda C', 'Tienda C', 'Tienda D', 'Tienda D'],
    'categoria': ['Ropa', 'Electrónica', 'Ropa', 'Hogar', 'Hogar', 'Electrónica', 'Ropa', 'Hogar'],
    'producto': ['Camisa', 'Celular', 'Pantalón', 'Lámpara', 'Sofá', 'Laptop', 'Zapatos', 'Mesa'],
    'ventas': [5000, 12000, 4000, 3000, 15000, 18000, 7000, 2000],
    'calificacion': [4.5, 4.7, 4.0, 3.8, 4.8, 4.9, 4.2, 3.5],
    'costo_envio': [20, 30, 25, 15, 40, 50, 22, 18]
}

df = pd.DataFrame(data)

# Mostrar los datos cargados
print("Datos cargados:")
print(df)

# 1. Análisis de facturación
facturacion = df.groupby('tienda')['ventas'].sum().reset_index()
facturacion = facturacion.sort_values('ventas', ascending=False)
print("\nFacturación total por tienda:")
print(facturacion)

# 2. Ventas por categoría
ventas_categoria = df.groupby(['tienda', 'categoria'])['ventas'].sum().reset_index()
print("\nVentas por categoría en cada tienda:")
print(ventas_categoria)

# 3. Clasificación promedio
evaluacion_prom = df.groupby('tienda')['calificacion'].mean().reset_index()
print("\nPromedio de evaluación por tienda:")
print(evaluacion_prom)

# 4. Productos más y menos vendidos
productos_ventas = df.groupby(['tienda', 'producto'])['ventas'].sum().reset_index()
productos_max = productos_ventas.loc[productos_ventas.groupby('tienda')['ventas'].idxmax()]
productos_min = productos_ventas.loc[productos_ventas.groupby('tienda')['ventas'].idxmin()]
print("\nProductos más vendidos por tienda:")
print(productos_max)
print("\nProductos menos vendidos por tienda:")
print(productos_min)

# 5. Promedio de envío
prom_envio = df.groupby('tienda')['costo_envio'].mean().reset_index()
print("\nCosto promedio de envío por tienda:")
print(prom_envio)

# Visualización de resultados

# Gráfico de Facturación Total por Tienda
plt.figure(figsize=(10,5))
sns.barplot(x='tienda', y='ventas', data=facturacion, palette='viridis')
plt.title("Facturación Total por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Facturación")
plt.xticks(rotation=45)
plt.show()

# Gráfico de Calificación Promedio por Tienda
plt.figure(figsize=(10,5))
sns.barplot(x='tienda', y='calificacion', data=evaluacion_prom, palette='coolwarm')
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.xticks(rotation=45)
plt.show()

# Gráfico de Costo Promedio de Envío por Tienda
plt.figure(figsize=(10,5))
sns.barplot(x='tienda', y='costo_envio', data=prom_envio, palette='magma')
plt.title("Costo Promedio de Envío por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo Promedio de Envío")
plt.xticks(rotation=45)
plt.show()

# Resumen de la tienda con menor rendimiento
tienda_menor = facturacion.sort_values('ventas').iloc[0]
print(f"\nLa tienda con menor rendimiento es {tienda_menor['tienda']} con una facturación de {tienda_menor['ventas']}")

```
